```python
# The above method is doing this by default
# did not remove the last row as it is a string
# axis can be set to index or columns
# index: drop rows if have missing values
# columns: drop columns if they have missing values
# how = any will drop rows with any missing values
# how = all will only drop if all the values are missing in the row
df_students.dropna(axis = 'index', how = 'all')
```

```python
# We need an email otherwise we will drop the row
# Note that how = all or any doesnt matter here
df_students.dropna(axis = 'index', how = 'any', subset = ['email'])
```

```python
# Dont drop rows which have last name or email
# how = all: in order to drop, both of the subsets should be missing
# use inplace = True for permanent change
df_students.dropna(axis = 'index', how = 'all', subset = ['last_name', 'email'])
```
